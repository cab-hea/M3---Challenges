```sql
/* 
Revenus par catégorie de produit : Calculez les revenus totaux générés pour chaque catégorie de produit.
(Tables : SalesOrderHeader, SalesOrderDetail, Product, ProductCategory)
Remarque : n'ayant pas de détails sur le calcul du revenu, je suis partie du principe qu'il correspondait au momtant de la colonne LineTotal pour chaque ligne de commande de la table SalesOrderDetail. Une autre possibilité, aurait pu être de tenir compte du prix de vente (ListPrice) et des coûts de chaque produit (StandardCost) ainsi que nombre d'articles vendus (OrderQty)
*/

Select 
  pc.Name as ProductCategory, 
  SUM(sod.LineTotal) as TotalRevenue
From productcategory as PC
 left join 
  Product p on p.ProductCategoryID = pc.ProductCategoryID
 left join 
  salesorderdetail sod on sod.ProductID = p.ProductID
Group by pc.Name
Order by TotalRevenue desc;
```
